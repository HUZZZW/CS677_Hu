Questions 3. which year was the lowest and which year was the highest?

```python
#3. which year was the lowest and which year was the highest?
job_title=[]
experience_level=[]
remote_ratio=[]
employee_residence=[]
with open('ds_salaries.csv',newline="") as Q3:
    reader = csv.DictReader(Q3)
    count_2020=count_2021=count_2022=0
    for row in reader:
        if (row["work_year"]=="2020"):
            count_2020+=int(row["salary_in_usd"])
        elif (row["work_year"]=="2021"):
            count_2021+=int(row["salary_in_usd"])
        elif (row["work_year"]=="2022"):
            count_2022+=int(row["salary_in_usd"])

        job_title.append(row["job_title"])
        experience_level.append(row["experience_level"])
        remote_ratio.append(row['remote_ratio'])
        employee_residence.append(row['employee_residence'])
        
    print("year 2020 salary:",count_2020)
    print("year 2021 salary:",count_2021)
    print("year 2022 salary:",count_2022)
    print("2021 was the highest year.")
```

Question 5. For which job title was it the minimum and maximum?

```python
# 5. For which job title was it the minimum and maximum?
job_title_totle_salary={}
job_title_totle_people={}
for name in job_title: 
    jb_count=0
    jb_human_count=0
    with open('ds_salaries.csv',newline="") as Q5:
        reader = csv.DictReader(Q5)
        for row in reader:
            #print(name)
            if (row["job_title"]==name): 
                jb_count=jb_count+int(row["salary_in_usd"])
                jb_human_count+=1
                #print(name,jb_count)
                job_title_totle_salary[name]=jb_count
                job_title_totle_people[name]=jb_human_count
        
print(job_title_totle_salary)
max_val=min_val=job_title_totle_salary['Data Science Manager']
max_res=min_res=""
for name in job_title:
    if (job_title_totle_salary[name]>max_val):
        max_res=name
    if (job_title_totle_salary[name]<min_val):
        min_res=name
print("the job title of the maximun salary is:",max_res,"\n","the job title of the minimun salary is:", min_res)
```
